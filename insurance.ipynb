```python
import pandas as pd

# Load the uploaded CSV file
file_path = "insurance.csv"
df = pd.read_csv(file_path)

# Display the first few rows of the dataset
df.head()


print(" Null Values:\n")
print(df.isnull().sum())

#  Check for Duplicate Rows
print("\n Duplicate Rows:\n")
print(f"Total Duplicates: {df.duplicated().sum()}")
# Show duplicate rows
duplicate_rows = df[df.duplicated()]
print(duplicate_rows)
# Remove duplicate rows
df = df.drop_duplicates()

# Confirm removal
print(f"Duplicates after removal: {df.duplicated().sum()}")


# Check Data Types
print("\n Data Types:\n")
print(df.dtypes)

```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set style
sns.set(style="whitegrid")
plt.figure(figsize=(10, 20))

# 1. Bar Chart: Average charges by region
plt.subplot(3, 2, 1)
region_charges = df.groupby("region")["charges"].mean().sort_values()
sns.barplot(x=region_charges.values, y=region_charges.index, palette="Blues_d")
plt.title("Average Charges by Region")
plt.xlabel("Average Charges")
plt.ylabel("Region")

# 2. Box Plot: Charges by Smoking Status
plt.subplot(3, 2, 2)
sns.boxplot(x="smoker", y="charges", data=df, palette="Set2")
plt.title("Charges by Smoking Status")

# 3. Histogram: BMI Distribution
plt.subplot(3, 2, 3)
sns.histplot(df["bmi"], bins=30, kde=True, color="skyblue")
plt.title("Distribution of BMI")

# 4. Scatter Plot: Age vs Charges colored by Smoker
plt.subplot(3, 2, 4)
sns.scatterplot(x="age", y="charges", hue="smoker", data=df, palette="coolwarm")
plt.title("Age vs Charges by Smoking Status")

# 5. Count Plot: Number of People per Region
plt.subplot(3, 2, 5)
sns.countplot(y="region", data=df, palette="viridis")
plt.title("Number of People per Region")

plt.tight_layout()
plt.show()
```
